```python
from pathlib import Path
import imagesize
import numpy as np
import seaborn as sns
```

```python
path_to_dataset = Path("bird_dataset")
suffix = "test_images_cropped"
```

```python
files = [x for x in path_to_dataset.glob("**/*") if x.is_file() and suffix in str(x)]
dimensions = np.array([imagesize.get(img_path) for img_path in files])[:, 0]
print(dimensions)
```

```python
sns.kdeplot(data=dimensions)
```
